# Min Max Scaler (Normalization)
* It basically converts values and normalizes them withing range 0 and 1.
* Usually we use Standardization for ML problems, but for Deep learning we must use Normalization.



```python
import seaborn as sns
```

```python
df = sns.load_dataset('taxis')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

`min_max.fit` will calculate the values

```python
min_max.fit(df[['distance','fare','tip']])
```

`min_max.transform` conert the values between 0 and 1 in dataset

```python
min_max.transform(df[['distance','fare','tip']])
```

```python
min_max.fit_transform(df[['distance','fare','tip']]) #Performs both operations at once
```
