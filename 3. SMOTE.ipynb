## SMOTE
SMOTE (Synthetic Minority Over-sampling Technique) is a machine learning technique that addresses class imbalance by generating synthetic instances for the minority class, enhancing model performance on underrepresented classes. It creates synthetic samples along line segments connecting existing minority class instances.

```python
from sklearn.datasets import make_classification
```

```python
# X Independent Feature 
# Y dependent feature

X,y = make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],random_state=1)
```

```python
import pandas as pd 
df1 = pd.DataFrame(X,columns = ['f1','f2'])
df2 = pd.DataFrame(y,columns = ['target'])
df = pd.concat([df1,df2],axis=1)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt 
plt.scatter(df['f1'],df['f2'],c=df['target'])
```

```python
from imblearn.over_sampling import SMOTE 
```

NOTE - SMOTE Technique creates more samples on place near to the datapoints,unlike upsampling which plots the data point on the same old datapoints available.

```python
##Transform the dataset 
oversample = SMOTE()
X,y = oversample.fit_resample(df[['f1','f2']],df['target'])
```
